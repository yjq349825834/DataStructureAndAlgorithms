```python
# Typically, no calculations are done until the base case is reached

def recur(n: int) -> int:
    """Recursion"""
    # Termination condition
    if n == 1:
        return 1
    # Recursive: recursive call
    res = recur(n - 1)
    # Return: return result
    return n + res


if __name__ == "__main__":
    print(recur(5))
```

recur(5) = 5 + recur(4)

recur(4) = 4 + recur(3)

recur(3) = 3 + recur(2)

recur(2) = 2 + recur(1)

recur(1) = 1  "reached base case"

recur(2) = 2 + 1

recur(3) = 3 + 2 + 1

recur(4) = 4 + 3 + 2 + 1

recur(5) = 5 + 4 + 3 + 2 + 1

```python
def tail_recur(n, res):
    """Tail recursion"""
    # Termination condition
    if n == 0:
        return res
    # Tail recursive call
    return tail_recur(n - 1, res + n)

if __name__ == "__main__":
    #
    print(tail_recur(10, 0))
```

tail_recur(5, 0) = tail_recur(4, 5)

tail_recur(4, 5) =  tail_recur(3, 5 + 4)

tail_recur(3, 5 + 4) =  tail_recur(2, 5 + 4 + 3)

tail_recur(2, 5 + 4 + 3) =  tail_recur(1, 5 + 4 + 3 + 2)

tail_recur(1, 5 + 4 + 3 + 2) = tail_recur(0, 5 + 4 + 3 + 2 + 1)
